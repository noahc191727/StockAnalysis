```python
import numpy as np
import pandas
import yfinance as yf
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split

```

```python
#Get the stock data
df = yf.download("AAPL", start= "2024-1-1", multi_level_index=False)
```

```python

df = df['Close']
df.head()
```

```python
# Variable for predicting n days out into the future
forecast_out = 30
#Create another column (the target or dependent variable) shifted 'n' units up
df['Prediction'] = df.shift(-forecast_out)
df.tail()
```

```python
### Create the independent data set (X)
# Convert the dataframe to a numpy array

X = np.array(df.drop(['Prediction']))
#Remove the last forecast_out rows
X2 = X.reshape(-1, 1)
X2 = X2[:-forecast_out]
X2
```

```python
### Create the dependent data set(y)
# Convert the dataffframe to a numpy array (All of the valuesd including the NaNs)
y = np.array(df['Prediction'])
y = y[:-forecast_out]

```

```python
# Split the data into 80 percent training and 20 percent testing
x_train, x_test, y_train, y_test = train_test_split(X2, y, test_size=0.2)
```

```python
# Create and train the Support Vector Machine (Regressor)
svr_rbf = SVR(kernel='rbf', C=1e3, gamma=0.1)
svr_rbf.fit(x_train, y_train)
```

```python
# Testing Model: Score returns the coefficient of determination, R^2 of the prediction
# The best possible score is 1.0
svm_confidence = svr_rbf.score(x_test, y_test)
svm_confidence
```

```python
# Create and train the Linear Regression Model
lr = LinearRegression()
# Train the model
lr.fit(x_train, y_train)
```

```python
lr_confidence = lr.score(x_test, y_test)
lr_confidence
```

```python
#Set x_forecast equal to the last 30 rows of the original data set from Close
x_forecast = np.array(df.drop(['Prediction']))[-forecast_out:]
x_forecast2 = x_forecast.reshape(-1, 1)
```

```python
#Print the predictions for the next n days
lr_prediction = lr.predict(x_forecast2)
lr_prediction

```
